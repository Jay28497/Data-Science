# Basic Pandas Operations

Pandas stands for panel data and is the core library for data manipulation and data analysis. It consist of single and multi-dimensional data structures for data manupulation.

```python
# import pandas
import pandas as pd
```

# Pandas Series Object

```python
# Series object is one-dimensional labeled array

s1 = pd.Series([1,2,3,4,5])
s1
```

```python
# Changing index

s1 = pd.Series([1,2,3,4,5], index=['a', 'b', 'c', 'd', 'e'])
s1
```

```python
# Series object from dictionary

s1 = pd.Series({'a':10, 'b':20, 'c':30})
s1
```

```python
type(s1)
```

```python
# Changin index position

s1 = pd.Series({'a':10, 'b':20, 'c':30}, index=['b', 'c', 'd', 'a'])
s1
```

# Extracting  Individual Element

```python
# Extracting a single element

s1 = pd.Series([1,2,3,4,5,6,7,8,9])
s1[3]
```

```python
# Extracting a sequence of element

s1[:4]
```

```python
# Extracting element from back

s1[-3:]
```

# Basic Math Operations On Series

```python
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([10,20,30,40,50])
```

```python
s1 + s2
```

```python
s1 + 5
```

```python
s1 - 5
```

```python
s1 * 5
```

```python
s1 / 10
```

# Pandas Dataframe

Dataframe is a 2 - dimensional labelled data structure. A dataframe comprises of rows and columns

```python
# Creating a dataframe

pd.DataFrame({"Name": ["Swagger", "Jay", "Alen"], "Marks": ["85", "80", "95"]})
```

# Dtaframe In-built Functions

```python
# Import and read csv file

iris = pd.read_csv("iris.csv")
```

```python
# Head -> show first 5 rows

iris.head()
```

```python
# Tail -> show last 5 rows

iris.tail()
```

```python
# Shape -> number of rows and cols know

iris.shape
```

```python
# Describe -> show general information about dataframe

iris.describe()
```

# .iloc[]

```python
# .iloc[startindex:endindex, startcol:endcol]

iris.iloc[5:11, 2:]
```

# .loc[]

```python
# .loc[startindex:endindex, ("specified col name", "specified col name")]

iris.loc[1:10, ("petal_length", "species")]
```

# Dropping  Colums

```python
iris.head()
```

```python
# axis = 1 for drop col

iris.drop("species", axis = 1)
```

# Dropping Rows

```python
iris.head()
```

```python
# axis = 0 for dorp row

iris.drop([5,6,7], axis = 0)
```

# Pandas Function

```python
iris.head()
```

```python
# Mean -> get mean value from the table column

iris.mean()
```

```python
# Min -> get min value from the table column

iris.min()
```

```python
iris[["sepal_width", "petal_width"]].apply(double_make)
```
