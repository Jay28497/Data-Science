# Basic Matplotlib Operatios 

Matplotlib is a python library used fordata visualization. You can create bar, plots, scatte-plots, histograms and a lot more with matplotlib.

```python
# Import matplotlib and numpy

import numpy as np
from matplotlib import pyplot as plt
```

# Line Plot

```python
x = np.arange(1, 11)
x
```

```python
y = 2 * x
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
# adding title and labels

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X - label")
plt.ylabel("Y - label")
plt.show()
```

```python
# Changing line aesthetics

plt.plot(x,y, color="g", linestyle=":", linewidth=2)
plt.show()
```

```python
# Adding two lines in the same plot

x = np.arange(1,11)
y1 = 2 * x
y2 = 3 * x
plt.plot(x, y1, color="g", linestyle=":", linewidth=2)
plt.plot(x, y2, color="r", linestyle="-", linewidth=3)
plt.title("Line Plot")
plt.xlabel("X - label")
plt.ylabel("Y - label")
plt.show()
```

```python
# Grid view

x = np.arange(1,11)
y1 = 2 * x
y2 = 3 * x
plt.plot(x, y1, color="g", linestyle=":", linewidth=2)
plt.plot(x, y2, color="r", linestyle="-", linewidth=3)
plt.title("Line Plot")
plt.xlabel("X - label")
plt.ylabel("Y - label")
plt.grid(True)
plt.show()
```

```python
# Adding sub- plots

x = np.arange(1,11)
y1 = 2 * x
y2 = 3 * x
plt.subplot(1,2,1)    # plt.subplot(row,col,index)
plt.plot(x, y1, color="b", linestyle=":", linewidth=2)
plt.subplot(1,2,2)
plt.plot(x, y1, color="y", linestyle=":", linewidth=2)
plt.show()
```

# Bar Plot

```python
student = {"jay": 75, "shraddha": 90, "alen": 53}
names = list(student.keys())
marks = list(student.values())
plt.bar(names, marks)
plt.show()
```

```python
# Horizontal bar plot

plt.barh(names, marks, color="g")
plt.show()
```

# Scatter Plot

```python
x = [10,20,30,40,50,60,70,80,90]
y = [8,1,7,2,0,3,7,3,2]
plt.scatter(x,y)
plt.show()
```

```python
# Changing mark aesthetics

plt.scatter(x,y, marker= "*", c="g", s=100)
plt.show()
```

```python
# Adding two markers in the same plot

x = [10,20,30,40,50,60,70,80,90]
a = [8,1,7,2,0,3,7,3,2]
b = [7,2,5,6,9,1,4,5,3]
plt.scatter(x,a, marker= "*", c="g", s=100)
plt.scatter(x,b, marker= ".", c="r", s=200)
plt.show()
```

```python
# Adding sub plots

x = [10,20,30,40,50,60,70,80,90]
a = [8,1,7,2,0,3,7,3,2]
b = [7,2,5,6,9,1,4,5,3]
plt.subplot(1,2,1)
plt.scatter(x,a, marker= "*", c="g", s=100)
plt.subplot(1,2,2)
plt.scatter(x,b, marker= ".", c="r", s=200)
plt.show()
```

# Histogram

```python
# Note: Normaly we can use numerical values

data = [1,3,3,3,3,9,9,5,4,4,8,8,8,6,7]
plt.hist(data)
plt.show()
```

```python
# use bins and changing aethetics

plt.hist(data, color="g", bins=4)
plt.show()
```

```python
# working with a dataset

# import pandas
import pandas as pd

#import and read csv file
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
plt.hist(iris["sepal_length"], color="r")
plt.show()
```

```python
plt.hist(iris["sepal_length"], bins=30, color="y")
plt.show()
```

# Box Plot

```python
a = [1,2,3,4,5,6,7,8,9]
b = [1,2,3,4,5,4,3,2,1]
c = [6,7,8,9,8,7,6,5,4]
data = list([a,b,c])
plt.boxplot(data)
plt.show()
```

# Violin Plot

```python
a = [1,2,3,4,5,6,7,8,9]
b = [1,2,3,4,5,4,3,2,1]
c = [6,7,8,9,8,7,6,5,4]
data = list([a,b,c])
plt.violinplot(data)
plt.show()
```

```python
a = [1,2,3,4,5,6,7,8,9]
b = [1,2,3,4,5,4,3,2,1]
c = [6,7,8,9,8,7,6,5,4]
data = list([a,b,c])
plt.violinplot(data, showmedians=True)
plt.show()
```

# Pie Chart

```python
fruits  = ["Apple", "Orange", "Mango", "Guava"]
quantity = [67,34,100,45]
plt.pie(quantity, labels=fruits)
plt.show()
```

```python
# Changing  aesthetics

plt.pie(quantity, labels=fruits, autopct="%0.1f%%", colors=["yellow","grey","blue","red"])
plt.show()
```

# DoughNut Chart

```python
fruits  = ["Apple", "Orange", "Mango", "Guava"]
quantity = [67,34,100,45]
plt.pie(quantity, labels=fruits, radius=2)
plt.pie([1], colors=["w"], radius=1)
plt.show()
```
